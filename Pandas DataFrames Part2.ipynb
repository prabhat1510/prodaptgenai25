# Pandas DataFrames - Null Values

```python
import numpy as np
import pandas as pd
d= {'A':[1,2, np.nan],
    'B':[5,np.nan, np.nan],
    'C':[1,2,3]
   }
d
```

```python
df = pd.DataFrame(d)
```

```python
df
```

```python
df.dropna() #axis=0
```

```python
df.dropna(axis=1)
```

```python
df.dropna(thresh=2)
```

```python
df.dropna(axis=1, thresh=2)
```

```python
df
```

```python
df.fillna("Null Values")
```

```python
df
```

```python
df
```

```python
df['A'].fillna(value = df['A'].mean(),inplace=True)
df

```

# Pandas DataFrame -  Groupby


```python
import numpy as np
import pandas as pd
data = {'Company': ['GOOG', 'GOOG', 'MSFT', 'MSFT','FB', 'FB'],
        'Person': ['Anu','Adhin','Diya', 'Rek','Viha','Krish'],
        'Sales':[200, 120,340, 124, 350, 650]
       }
df = pd.DataFrame(data)
df
```

```python
company_group = df.groupby("Company")
print(company_group)
```

```python
company_group.sum()
```

```python
company_group.sum().loc['FB']
```

```python
df.groupby("Company").sum().loc["FB"]
```

```python
df.groupby("Company").count()
```

```python
#Displays person name as well, python displays the max string in the person section as well
df.groupby("Company").max()
```

```python
df.groupby("Company").min()
```

```python
df.groupby("Company").describe()
```

# DataFrames - Joins

```python
import numpy as np
import pandas as pd
dic1 ={'A':['A0','A1','A2'],
       'B': ['B0','B1','B2'],
       'index':['K0','K1','K2']
      }
dic2 = {'C':['C0','C1','C2'],
       'D': ['D0','D1','D2'],
       'index':['K0','K2','K3']
      }

left = pd.DataFrame(dic1)
right = pd.DataFrame(dic2)
```

```python
print(left)
print(right)
```

```python
pd.merge(left,right,how="inner",on="index")
```

```python
pd.merge(left,right,how="inner", on="index")['C']
```

```python
pd.merge(left,right,how="left", on="index")
```

```python
pd.merge(left,right,how="right", on="index")
```

```python
pd.concat([left,right])
```

```python
pd.concat([left,right], axis=1)
```

# DataFrames - unique, value_counts, sort_values, isnull, lambda, Pivot

```python
import numpy as np
import pandas as pd
df= pd.DataFrame({
    'col1':[1,2,3,4],
    'col2':[444,555,666,444],
    'col3':['abc','def','ghi','xyz']
})
df
```

```python
df['col2'].unique
```

```python
df['col2'].value_counts()
```

```python
df.sort_values('col2')
```

```python
df.isnull()
```

```python
def times2(x):
    return x*2
df['col1'].apply(times2)
```

```python
df['col2'].apply(lambda x:x*2)
```

```python
data = {
    'A' :['foo', 'foo', 'foo','bar','bar','bar'],
    'B' :['one','one','two','two','one','one'],
    'C' :['x','y','x','y','x','y'],
    'D' :[1,3,2,5,4,1]
}
df = pd.DataFrame(data)
df
```
